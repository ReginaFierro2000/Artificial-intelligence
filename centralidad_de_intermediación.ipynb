## Análisis

¿Qué se puede decir de los vértices de mayor centralidad? ¿En qué cambian? 

Los vértices con la centralidad de intermediación más alta son los nodos más importantes de una red en términos de control del flujo de información o recursos entre otros nodos. Específicamente, son los nodos que se encuentran en la mayor cantidad de caminos más cortos entre todos los demás pares de nodos en la red. Estos nodos pueden actuar como puentes entre otros nodos o comunidades, y su eliminación de la red puede afectar significativamente la conectividad y la eficiencia generales de la red. En las redes sociales, los vértices con alta centralidad de intermediación a menudo se consideran "influenciadores" potenciales que pueden impactar las opiniones o comportamientos de otros en la red.

Por esas razones nos podemos dar cuenta que en el primer grafo uno de los nodos más importantes es el de Thalía(0.5476923076923077), aunque no tiene muchas conexiones pero se observa que tiene la mayor cantidad de caminos más cortos y si no existiera ese nodo rompería la comunicación entre cuatro comunidades de artistas. En cambio se observa que en el segundo grafo George Fox(0.47569259800029035) que es el más popular de los cuaqueros sigue teniendo mayor peso en el segundo grafo, significa que con su pérdida de nodo rompería completamente las comunidades que se formaron en su Sociedad Religiosa.  

Algunas propiedades clave de los vértices con alta centralidad de intermediación son:

  1. Desempeñan un papel fundamental en la estructura general del gráfico, ya que su eliminación puede aumentar significativamente el diámetro del gráfico o desconectarlo.

  2. Es probable que sean importantes para facilitar el flujo de información o recursos en el gráfico, ya que se encuentran en muchas de las rutas más eficientes entre las diferentes partes del gráfico.

  3. Pueden ser vulnerables a ataques o interrupciones, ya que su eliminación puede tener un impacto desproporcionadamente grande en la conectividad o la funcionalidad del gráfico.

  4. Se pueden utilizar como objetivos de intervención u optimización en diversas aplicaciones, como algoritmos de enrutamiento, análisis de redes sociales o planificación del transporte.

¿Cómo se comparan con los vértices obtenidos en la actividad de centralidad de grado, de eigenvector, de Katz y PageRank?

El grado de centralidad es una medida del número de conexiones que tiene un nodo en una red. En otras palabras, cuenta el número de aristas que están conectadas a un nodo en particular.

La centralidad del vector propio es una medida de la importancia de un nodo en una red. Tiene en cuenta el número y la calidad de las conexiones que tiene un nodo y también la calidad de las conexiones de sus vecinos. Los nodos con alta centralidad de vector propio son importantes porque están conectados a otros nodos importantes.

La centralidad de Katz es una medida de la influencia de un nodo en una red. Es similar a la centralidad del vector propio, pero también tiene en cuenta la distancia entre los nodos. Los nodos que están más cerca unos de otros se consideran más influyentes.

PageRank es una medida de la importancia de un nodo en una red basada en la idea de que es probable que los nodos importantes estén vinculados a otros nodos importantes. Se utiliza para clasificar las páginas web en los motores de búsqueda.

La centralidad de intermediación es una medida de la importancia de un nodo en una red basada en la cantidad de veces que aparece en el camino más corto entre otros nodos. Los nodos con alta centralidad de intermediación juegan un papel importante en la conexión de diferentes partes de la red y, a menudo, son críticos para el flujo de información.
