##Price

Para generar y visualizar redes usando el método de Price, podemos usar la biblioteca NetworkX en Python. Aquí hay un ejemplo que genera una red con 100 nodos y la visualiza usando un gráfico de red y un histograma de la distribución de grados:


```python
import networkx as nx
import random
import matplotlib.pyplot as plt

# Definir el número de nodos y gráfico inicial
n = 100
G = nx.Graph()
G.add_node(0)

# se agrega los nodos restantes usando el algoritmo de Price
for i in range(1, n):
    G.add_node(i)
    # Se selecciona un nodo aleatorio para conectarse
    target = random.choice(list(G.nodes()))
    # Se agrega un borde al nodo seleccionado
    G.add_edge(i, target)

# Se crea el gráfico de la red
nx.draw(G, with_labels=True)
plt.show()

#Se traza el histograma de distribución de grados
degrees = [G.degree(n) for n in G.nodes()]
plt.hist(degrees, bins=range(n))
plt.xlabel('Grado')
plt.ylabel('Número de nodos')
plt.show()
```

Este código primero inicializa un gráfico no dirigido vacío y agrega un solo nodo. Luego agrega iterativamente los nodos restantes utilizando el algoritmo de Price: para cada nuevo nodo, se elige un nodo existente al azar y se crea un nuevo borde entre el nuevo nodo y el nodo existente. Finalmente, el código usa NetworkX para crear un gráfico de red de la red resultante y traza un histograma de la distribución de grados.

El modelo de Price se usa a menudo para modelar el crecimiento de las redes libres de escala del mundo real, como las redes sociales, las redes de citas e Internet.

##Barabasi-Albert

Estos son los pasos para la implementación del método de Barabasi-Albert:

  1.Definir el número inicial de nodos y el número de nodos que se añadirán en cada iteración.

  2.Generar la red utilizando la función barabasi_albert_graph() de la librería NetworkX.

  3.Visualizar la red utilizando la función draw() de NetworkX.

  4.Calcular la distribución de grados de la red utilizando la función degree_histogram() de NetworkX.

  5.Visualizar la distribución de grados utilizando un histograma.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Definimos el número inicial de nodos y el número de nodos que se añadirán en cada iteración
num_nodos_inicial = 10
num_nodos_nuevos = 2

# Generamos la red utilizando el método Barabasi-Albert
red = nx.barabasi_albert_graph(num_nodos_inicial, num_nodos_nuevos)

# Visualizamos la red
nx.draw(red, with_labels=True)
plt.show()

# Calculamos la distribución de grados
distribucion_grados = nx.degree_histogram(red)

# Visualizamos la distribución de grados
plt.hist(distribucion_grados)
plt.xlabel('Grado')
plt.ylabel('Número de nodos')
plt.show()
```

Este código generó una red utilizando el método de Barabasi-Albert con un número inicial de 10 nodos y 2 nodos nuevos añadidos en cada iteración. También se calculó y mostró un histograma de la distribución de grados de la red.

Los modelos de Price y Barabási-Albert se aplican a menudo en el estudio de redes complejas y su evolución a lo largo del tiempo. Estos modelos son especialmente útiles para modelar redes reales como redes sociales, redes de transporte, redes de energía, redes de información y redes de financiamiento.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Definimos el número inicial de nodos y el número de nodos que se añadirán en cada iteración
num_nodos_inicial = 25
num_nodos_nuevos = 2

# Generamos la red utilizando el método Barabasi-Albert
red = nx.barabasi_albert_graph(num_nodos_inicial, num_nodos_nuevos)

# Visualizamos la red
nx.draw(red, with_labels=True)
plt.show()

# Calculamos la distribución de grados
distribucion_grados = nx.degree_histogram(red)

# Visualizamos la distribución de grados
plt.hist(distribucion_grados)
plt.xlabel('Grado')
plt.ylabel('Número de nodos')
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Definimos el número inicial de nodos y el número de nodos que se añadirán en cada iteración
num_nodos_inicial = 25
num_nodos_nuevos = 6

# Generamos la red utilizando el método Barabasi-Albert
red = nx.barabasi_albert_graph(num_nodos_inicial, num_nodos_nuevos)

# Visualizamos la red
nx.draw(red, with_labels=True)
plt.show()

# Calculamos la distribución de grados
distribucion_grados = nx.degree_histogram(red)

# Visualizamos la distribución de grados
plt.hist(distribucion_grados)
plt.xlabel('Grado')
plt.ylabel('Número de nodos')
plt.show()
```

El modelo de Price funciona añadiendo gradualmente nuevos nodos a la red y conectándolos al nodo existente con mayor grado. Esto crea una red que tiene una distribución de grado exponencial, lo que significa que la mayoría de los nodos tienen un grado bajo, y sólo unos pocos nodos tienen un grado alto.

El modelo de Barabási-Albert, por otro lado, funciona añadiendo gradualmente nuevos nodos a la red y conectándolos a los nodos existentes con una probabilidad proporcional al grado del nodo. Esto crea una red que tiene una distribución de grado libre de escala, lo que significa que hay muchos nodos con un grado bajo y unos pocos nodos con un grado muy alto.

En particular, el modelo de Barabási-Albert se utiliza comúnmente para modelar el crecimiento de redes complejas en las que algunos nodos parecen tener una mayor probabilidad de recibir nuevos enlaces, lo que conduce a la aparición de nodos altamente conectados o "súper-nodos". Además, el modelo de Price y el modelo de Barabási-Albert se pueden utilizar para analizar las propiedades topológicas y estructurales de las redes complejas, como la distribución de grados, la red de enrutamiento y la conectividad.

En resumen, los modelos de Price y Barabási-Albert son ampliamente utilizados en la ciencia de redes para estudiar la evolución y estructura de las redes complejas en diversos campos de aplicación.
