##Modelo de Watts-Strogatz


Para generar y visualizar una red usando el método de Watts-Strogatz, se pueden seguir los siguientes pasos:

  1.Definir los parámetros de la red: número de nodos, grado medio, y probabilidad de reconexión.

  2.Generar una red regular de nodos, en la que cada nodo está conectado a sus c vecinos más cercanos a la izquierda y a la derecha. Esto se puede lograr creando un anillo y conectando cada nodo con sus k vecinos más cercanos.

  3.Para cada conexión en la red, con una probabilidad p, reconectar el nodo con otro nodo aleatorio en la red. Esto se puede lograr generando un número aleatorio y si es menor que p, reconectar la conexión.

  4.Visualizar la red utilizando una librería de graficación.

  5.Calcular el histograma de distribución de grados y el coeficiente de agrupamiento.

  6.Calcular las geodésicas promedio utilizando un algoritmo de búsqueda de caminos cortos, como el algoritmo de Dijkstra.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Parámetros de la red
n = 100  # número de nodos
k = 4  # grado medio
p = 0.2  # probabilidad de reconexión

# Generar una red regular
G = nx.watts_strogatz_graph(n, k, p)

# Visualizar la red
pos = nx.circular_layout(G)
nx.draw(G, with_labels=True)
plt.show()

# Visualiza la red de otra forma (en circunferencia)
pos = nx.circular_layout(G)
nx.draw(G,pos, with_labels=True)
plt.show()

# Calcular el histograma de distribución de grados
grados = [d for n, d in G.degree()]
plt.hist(grados, bins='auto')
plt.xlabel('Grado')
plt.ylabel('Frecuencia')
plt.show()

# Calcular el coeficiente de agrupamiento
c = nx.average_clustering(G)
print("Coeficiente de agrupamiento:", c)

# Calcular las geodésicas promedio
d = nx.average_shortest_path_length(G)
print("Geodésicas promedio:", d)
```

La red Watts-Strogatz se genera mediante la función nx.watts_strogatz_graph(), que toma n nodos, k grado de cada nodo y p probabilidad de recableado. Luego, la red se visualiza utilizando la función nx.draw().

La distribución de grados se calcula usando la función nx.degree_histogram(), que devuelve la frecuencia de cada grado en la red. El histograma se traza utilizando la función plt.bar().

El coeficiente de agrupamiento se calcula mediante la función nx.average_clustering, que devuelve un diccionario de coeficientes de agrupamiento para cada nodo de la red. Luego, el coeficiente de agrupamiento promedio se calcula tomando la suma de todos los coeficientes de agrupamiento dividida por el número de nodos en la red.

La longitud promedio de la ruta más corta se calcula mediante la función nx.shortest_path_length(), que devuelve la longitud de la ruta más corta entre cada par de nodos de la red. A continuación, se calcula la longitud media de la ruta más corta tomando la suma de todas las longitudes de la ruta más corta dividida por el número total de pares de nodos en la red.

Ahora se probará con distintos parámetros.

El modelo de Watts-Strogatz es un tipo de modelo de gráfico aleatorio desarrollado por Duncan J. Watts y Steven H. Strogatz en 1998.

El modelo se caracteriza por su capacidad para generar redes que tienen un alto grado de agrupamiento y una longitud de ruta promedio corta. Hace esto comenzando con un anillo de n nodos, donde cada nodo está conectado a sus c vecinos más cercanos, con c/2 en cada lado. Luego, con probabilidad p, cada borde de la red se vuelve a cablear, conectando el nodo a un nuevo nodo aleatorio en la red, creando atajos.

A través de este proceso, Watts y Strogatz pudieron demostrar que las redes del mundo pequeño se pueden generar a partir de un modelo simple e intuitivo diseñado para comprender las propiedades de las redes del mundo real. Desde entonces, el modelo de Watts-Strogatz se ha convertido en uno de los modelos más utilizados para comprender y analizar redes de mundo pequeño en varios campos de investigación, incluida la ciencia de redes, la sociología, la informática y las matemáticas.

Una aplicación práctica del modelo de Watts-Strogatz es en el diseño y análisis de redes de comunicación. Por ejemplo, se ha utilizado para estudiar la difusión de información o enfermedades a través de las redes sociales o para optimizar el enrutamiento de datos a través de Internet. Además, se ha utilizado para estudiar la estructura y función de las redes neuronales en el cerebro humano. En general, el modelo ha demostrado ser una herramienta valiosa para comprender la naturaleza compleja de las redes del mundo real.
