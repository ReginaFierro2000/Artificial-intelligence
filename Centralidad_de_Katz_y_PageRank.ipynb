##Análisis

¿Qué se puede decir de los vértices de mayor centralidad? ¿En qué cambian? 

Tanto la centralidad de Katz como el algoritmo PageRank se utilizan para medir la importancia de los nodos en una red. En el primer grafo y en el segundo grafo se implementó ambos algoritmos dando diferentes valores, el más alto fue Pitbull (primer grafo) y George Fox (segundo grafo).

La centralidad de Katz mide la influencia relativa de un nodo dentro de una red midiendo la suma de las contribuciones de todos los demás nodos de la red a ese nodo. Se basa en la idea de que un nodo es importante si está conectado a otros nodos importantes.

El algoritmo PageRank, por otro lado, mide la importancia de un nodo en función del número y la calidad de sus enlaces entrantes. Los motores de búsqueda lo utilizan para clasificar las páginas web en sus resultados de búsqueda.

Si bien ambos algoritmos son similares en el sentido de que miden la importancia de los nodos, difieren en su enfoque y en los tipos de redes para los que son más adecuados. La centralidad de Katz se usa mejor en redes más pequeñas donde hay muchas conexiones entre nodos y para el primer grafo es más favorable por el tamaño del grafo, mientras que el algoritmo PageRank es más adecuado para redes más grandes como el segundo grafo de los cuaqueros.

En general, ambos algoritmos tienen sus puntos fuertes y débiles y pueden ser útiles en diferentes escenarios de análisis de red.

1. La centralidad de Katz tiene en cuenta la influencia de los vecinos de un nodo, mientras que el algoritmo PageRank considera la importancia de los nodos que apuntan a él.
2. La centralidad de Katz utiliza una función de decaimiento exponencial para ponderar rutas de diferentes longitudes, mientras que el algoritmo PageRank pondera los enlaces por igual.
3. La centralidad de Katz requiere la elección de un parámetro alfa que determina la importancia relativa de las rutas cortas y largas, mientras que el algoritmo PageRank no requiere ningún parámetro adicional.

En el caso de la centralidad de Kats se puede observar que el segundo grafo tuvo mayor valor en el vertice de mayor centralidad que del primer grafo, esto quiere decir que George Fox tuvo mayor influencia que Pitbull pero se sabe que la cantidad de nodos del primer grafo son de 27 y del segundo grafo son de 67 y los resultados obtenidos de la centralidad de Kats no varian mucho (Node Pitbull: Katz centrality = 0.30609802750368764 y Node George Fox: Katz centrality =0.3745791738369515) varian por muy pocos decimales.

En el caso de la centralidad de PageRank fue lo contrario, se observa mayor peso en el primer grafo con Pitbull: PageRank centrality =       0.15380455933785933 y el segundo grafo con George Fox: PageRank centrality =0.08796231170826019. 

Como resultado, los vértices de mayor centralidad en un gráfico pueden diferir según el algoritmo utilizado. En general, la centralidad de Katz tiende a favorecer los vértices con muchos vecinos cercanos, mientras que el algoritmo PageRank tiende a favorecer los vértices con muchos enlaces entrantes de alta calidad. Sin embargo, en algunos casos, los dos algoritmos pueden producir resultados similares, especialmente si el gráfico tiene una estructura relativamente uniforme.

¿Cómo se comparan con los vértices obtenidos en la actividad de centralidad de grado y de eigenvector?

La centralidad de grado, la centralidad de vector propio, la centralidad de Katz y PageRank son todas medidas de centralidad en una red, pero difieren en la forma en que calculan la importancia de un nodo en la red.

La centralidad de grado es simplemente la cantidad de bordes o conexiones que tiene un nodo con otros nodos en la red. Los nodos con un mayor grado de centralidad están más conectados con otros nodos y pueden ser más importantes en la red.

La centralidad del vector propio tiene en cuenta no solo el número de conexiones que tiene un nodo, sino también la centralidad de los nodos a los que está conectado. Un nodo con alta centralidad de vector propio está conectado a otros nodos que también son importantes en la red.

La centralidad de Katz es similar a la centralidad del vector propio, pero también considera caminos entre nodos que no están conectados directamente. Asigna puntajes de centralidad más altos a los nodos que forman parte de rutas más largas entre nodos importantes en la red.

PageRank asigna puntajes más altos a los nodos que tienen muchos enlaces entrantes de otros nodos importantes.

En general, cada una de estas medidas de centralidad captura un aspecto diferente de la importancia de un nodo en la red, y la elección de cuál usar puede depender del contexto específico de la red que se analiza.
