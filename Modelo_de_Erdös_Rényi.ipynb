##Gráficas aleatorias

Para generar una red aleatoria usando el método de gráficas aleatorias, podemos seguir los siguientes pasos:

  1.Definir el número de nodos que deseamos tener en la red. Esto puede ser un número fijo o aleatorio, dependiendo del problema que queramos resolver.

  2.Generar una matriz de adyacencia de tamaño NxN, donde N es el número de nodos. Esta matriz contendrá información sobre las conexiones entre los nodos de la red. Inicialmente, todas las entradas de la matriz se establecen en cero.

  3.Para cada par de nodos (i,j), con i<j, generamos un número aleatorio entre 0 y 1. Si este número es menor o igual a una probabilidad de conexión p, establecemos la entrada (i,j) de la matriz de adyacencia en 1 (indicando una conexión entre los nodos i y j). De lo contrario, la entrada se mantiene en cero.

  4.Usando la matriz de adyacencia generada, podemos construir un objeto de red. Para hacerlo, podemos utilizar una librería de visualización de gráficas, como NetworkX, y agregar nodos y conexiones a la red a través de la matriz de adyacencia.

  5.Visualizar la red generada usando una herramienta de visualización de gráficas, como Matplotlib o Gephi.

```python
import random
import networkx as nx
import matplotlib.pyplot as plt

# Definimos el número de nodos
num_nodos = 10

# Generamos la matriz de adyacencia
matriz_ady = [[0 for i in range(num_nodos)] for j in range(num_nodos)]
for i in range(num_nodos):
    for j in range(i+1, num_nodos):
        prob_conexion = random.random()
        if prob_conexion <= 0.5: 
            matriz_ady[i][j] = 1
            matriz_ady[j][i] = 1

# Creamos el objeto de red y agregamos nodos y conexiones
red = nx.Graph()
for i in range(num_nodos):
    red.add_node(i)
for i in range(num_nodos):
    for j in range(i+1, num_nodos):
        if matriz_ady[i][j] == 1:
            red.add_edge(i, j)

# Visualizamos la red
nx.draw(red, with_labels=True)
plt.show()

# Calculamos la distribución de grados
distribucion_grados = nx.degree_histogram(red)

# Visualizamos la distribución de grados
plt.hist(distribucion_grados)
plt.xlabel('Grado')
plt.ylabel('Número de nodos')
plt.show()
```

Este código generará una red aleatoria con 10 nodos y una probabilidad de conexión de 0.5.

##Erdös-Rényi

Para generar y visualizar redes utilizando el método de gráficas aleatorias y Erdös-Rényi, podemos seguir los siguientes pasos:

  1.Definir el número de nodos y la probabilidad de conexión para la red.

  2.Generar la red utilizando la función erdos_renyi_graph() de la librería NetworkX.

  3.Visualizar la red utilizando la función draw() de NetworkX.

  4.Calcular la distribución de grados de la red utilizando la función degree_histogram() de NetworkX.

  5.Visualizar la distribución de grados utilizando un histograma.


```python
import networkx as nx
import matplotlib.pyplot as plt

# Definimos el número de nodos y la probabilidad de conexión
num_nodos = 100
prob_conexion = 0.05

# Generamos la red utilizando Erdös-Rényi
red = nx.erdos_renyi_graph(num_nodos, prob_conexion)

# Visualizamos la red
nx.draw(red, with_labels=True)
plt.show()

# Calculamos la distribución de grados
distribucion_grados = nx.degree_histogram(red)

# Visualizamos la distribución de grados
plt.hist(distribucion_grados)
plt.xlabel('Grado')
plt.ylabel('Número de nodos')
plt.show()
```

```python
# Definimos el número de nodos y la probabilidad de conexión
num_nodos = 100
prob_conexion = 0.5

# Generamos la red utilizando Erdös-Rényi
red = nx.erdos_renyi_graph(num_nodos, prob_conexion)

# Visualizamos la red
nx.draw(red, with_labels=True)
plt.show()

# Calculamos la distribución de grados
distribucion_grados = nx.degree_histogram(red)

# Visualizamos la distribución de grados
plt.hist(distribucion_grados)
plt.xlabel('Grado')
plt.ylabel('Número de nodos')
plt.show()
```

Este código generó una red aleatoria utilizando el método de Erdös-Rényi con 100 nodos y una probabilidad de conexión de 0.05 y 0.5.También se calculó y mostró un histograma de la distribución de grados de la red.

Para ambos modelos entre mayor sea la probabilidad es muy probable que cada nodo comparta una arista para cada uno de los nodos que se encuentran en el grafo y entre menor sea el grado aumenta el número de nodos.

El modelo de redes aleatorias, también conocido como modelo de Erdös-Rényi, es uno de los modelos más simples y utilizados en el estudio de redes complejas. En este modelo, los nodos se unen aleatoriamente con una probabilidad constante p. Como resultado, la distribución de grado sigue una distribución binomial, lo que significa que la mayoría de los nodos tienen un grado similar y pocos nodos tienen un grado muy alto o muy bajo.

Este modelo de red aleatoria aporta valiosos puntos de referencia para otros modelos más complejos, ya que exhibe propiedades comunes en redes complejas reales, como el coeficiente de clustering y la longitud de camino promedio.

Sin embargo, las redes generadas por este modelo difieren significativamente de las redes complejas reales, ya que la mayoría de las redes en el mundo real muestran una distribución de grado libre de escala. Por lo tanto, su relevancia puede ser limitada en estudios que se centran en propiedades específicas de la red, como la robustez ante fallos o la propagación de la información.

La distribución de grados es una medida importante en teoría de redes, ya que puede indicar características como la centralidad de los nodos y la presencia de nodos altamente conectados (hubs). Las redes generadas con el método de Erdös-Rényi se aplican en una amplia variedad de campos, incluyendo biología, informática, física, ciencias sociales y económicas, y muchos otros.
